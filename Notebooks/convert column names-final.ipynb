```python
import pandas as pd
```

```python
birth_data_2 = pd.read_csv('../data/birth_data_final.csv')
```

```python
birth_data_2
```

```python
birth_data_2['Place'] = birth_data_2['Place'].replace({1: 'hospital', 2: 'birth_center'})
```

```python
birth_data_2['Race'] = birth_data_2['Race'].replace({1: 'white', 2: 'black', 4: 'Asian'})
```

```python
birth_data_2['Delivery_Type'] = birth_data_2['Delivery_Type'].replace({1: 'Vaginal', 2: 'C-Section', 9: 'Unknown'})
```

```python
birth_data_2['attend'] = birth_data_2['attend'].replace({1: 'MD', 3: 'CNM', 4: 'Midwife'})
```

```python
birth_data_2['Payment_Type'] = birth_data_2['Payment_Type'].replace({1: 'Medicaid', 2: 'Private_Ins', 3: 'Self_Pay', 4: 'Other'})
```

```python
birth_data_2['Mother_Age'] = birth_data_2['Mother_Age'].replace({1: 'under_15yrs', 2: '15_19yrs', 3: '20_24yrs', 4: '25-29yrs', 5: '30-34yrs', 6: '35-39yrs', 7: '40-44yrs', 8: '45-49yrs', 9: '50-54yrs'})
```

```python
birth_data_2['Edu_level'] = birth_data_2['Edu_level'].replace({1: '8th_grade_or_less', 2: '9th-12th_no_diploma', 3: 'HS_grad_or_GED', 4: 'some_college_credit_no_degree', 5: 'AA, AS', 6: 'Bachelors', 7: 'Masters', 8: 'Doctorate', 9: 'Unknown'})
                                
```

```python
birth_data_2.to_csv('birth_data_2.csv', index=False)
```
