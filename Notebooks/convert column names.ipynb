```python
import pandas as pd
```

```python
birth_data_1 = pd.read_csv('../final data/birth_data.csv')
```

```python
birth_data_1
```

```python
birth_data_1['Place'] = birth_data_1['Place'].replace({1: 'hospital', 2: 'birth_center'})
```

```python
birth_data_1['Race'] = birth_data_1['Race'].replace({1: 'white', 2: 'black', 4: 'Asian'})
```

```python
birth_data_1['Delivery_Type'] = birth_data_1['Delivery_Type'].replace({1: 'Vaginal', 2: 'C-Section', 9: 'Unknown'})
```

```python
birth_data_1['attend'] = birth_data_1['attend'].replace({1: 'MD', 3: 'CNM', 4: 'Midwife'})
```

```python
birth_data_1['Payment_Type'] = birth_data_1['Payment_Type'].replace({1: 'Medicaid', 2: 'Private_Ins', 3: 'Self_Pay', 4: 'Other'})
```

```python
birth_data_1
```

```python
birth_data_1.to_csv('birth_data_1.csv', index=False)
```
